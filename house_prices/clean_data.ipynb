```python
combined['MSSubClass'] = combined['MSSubClass'].astype('category')
combined['OverallQual'] = combined['OverallQual'].astype('category')
combined['OverallCond'] = combined['OverallCond'].astype('category')
combined['YearBuilt'] = combined['YearBuilt'].astype('category')
combined['YearRemodAdd'] = combined['YearRemodAdd'].astype('category')
combined['GarageYrBlt'] = combined['GarageYrBlt'].astype('category')
combined['MoSold'] = combined['MoSold'].astype('category')
combined['YrSold'] = combined['YrSold'].astype('category')
combined['GarageCars'] = combined['GarageCars'].astype('category')
combined['Fireplaces'] = combined['Fireplaces'].astype('category')
combined['TotRmsAbvGrd'] = combined['TotRmsAbvGrd'].astype('category')
combined['KitchenAbvGr'] = combined['KitchenAbvGr'].astype('category')
combined['BedroomAbvGr'] = combined['BedroomAbvGr'].astype('category')
combined['HalfBath'] = combined['HalfBath'].astype('category')
combined['FullBath'] = combined['FullBath'].astype('category')
combined['BsmtHalfBath'] = combined['BsmtHalfBath'].astype('category')
combined['BsmtFullBath'] = combined['BsmtFullBath'].astype('category')
combined['BedroomAbvGr'] = combined['BedroomAbvGr'].astype('category')
```

```python
objects = combined.select_dtypes(include='object')
cols = objects.columns
for i in cols:
    combined[i] = combined[i].astype('category')
```

```python
combined.info()
```

```python
combined['bathrooms'] = (combined['HalfBath'].astype('int') * 0.5 + combined['FullBath'].astype('int') \
                    + combined['BsmtFullBath'].astype('int') + combined['BsmtHalfBath'].astype('int') * 0.5 \
                     ).astype('category')

combined['total_sq_ft'] = combined['BsmtFinSF1'] + combined['BsmtFinSF2'] + combined['1stFlrSF'] + combined['2ndFlrSF']

combined['total_sq_ft_plus_unf'] = combined['BsmtFinSF1'] + combined['BsmtFinSF2'] + combined['1stFlrSF'] + combined['2ndFlrSF'] \
                                + combined['BsmtUnfSF']
    
def remod(row):
    remod = 0
    if row['YearRemodAdd'] > row['YearBuilt']:
        remod = 1
    return remod

combined['remodel'] = combined.apply(remod, axis=1)

combined['total_porch_sf'] = combined['WoodDeckSF'] + combined['OpenPorchSF'] + combined['EnclosedPorch'] + combined['3SsnPorch'] + \
                          combined['ScreenPorch']
```

```python
combined.to_csv('combined.csv', index=False)
```
