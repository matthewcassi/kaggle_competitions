```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train_clean.csv')
res = pd.read_csv('cleaned_resources.csv')
```

```python
train.head()
```

```python
train['project_submitted_datetime'] = pd.to_datetime(train['project_submitted_datetime'])
```

```python
train = train.drop(
    ['Unnamed: 0',
     'quantity',
     'total_price',
     'min_quantity',
     'max_quantity',
     'min_price',
     'max_price',
     'description',
     'tag_text',
     'project_essay_1', 
     'project_essay_2', 
     'project_essay_3', 
     'project_essay_4', 
     'project_resource_summary'], 1)
```

```python
train.columns
```

```python
res.head()
```

```python
res.isna().any()
```

```python
train = train.merge(res, left_on='id', right_on='id', how='left')
```

```python
train.head()
```

```python
train['hour'] = train['project_submitted_datetime'].dt.hour
train['minute'] = train['project_submitted_datetime'].dt.minute
```

```python
train.head()
```

```python
train.columns
```

```python
train = train.drop('year', 1)
```

```python
train.drop('project_title', 1, inplace=True)
```
