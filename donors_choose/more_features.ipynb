```python
import pandas as pd
import numpy as np
```

```python
train1 = pd.read_csv('train.csv')
```

```python
teacher = pd.pivot_table(train1, index='teacher_id', columns='project_is_approved', values='id', aggfunc=len)
teacher = teacher.fillna(0)
```

```python
teacher['per_approved'] = (teacher[1]) / (teacher[0] + teacher[1])
```

```python
train1['project_essay_3'] = train1['project_essay_3'].fillna('')
train1['project_essay_4'] = train1['project_essay_4'].fillna('')
```

```python
train1['title_exc'] = train1['project_title'].str.count(r'!')
train1['essay1_exc'] = train1['project_essay_1'].str.count(r'!')
train1['essay2_exc'] = train1['project_essay_2'].str.count(r'!')
train1['essay3_exc'] = train1['project_essay_3'].str.count(r'!')
train1['essay4_exc'] = train1['project_essay_4'].str.count(r'!')
train1['resource_exc'] = train1['project_resource_summary'].str.count(r'!')
train1['essay3_exc'] = train1['essay3_exc'].fillna('0')
train1['essay4_exc'] = train1['essay4_exc'].fillna('0')
```

```python
train1['total_exc'] = train1['title_exc'] + train1['essay1_exc'] + train1['essay2_exc'] + train1['essay3_exc'] + train1['essay4_exc'] + train1['resource_exc']
train1['total_exc'] = train1['total_exc'].fillna('0')
```

```python
teacher[['per_approved']].to_csv('approval_prop.csv')
```
